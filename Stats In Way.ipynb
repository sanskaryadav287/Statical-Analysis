###   Median Methods   ###

```python
###   Median  ###
data=[21,23,22,33,34,43,33,44,45,54,55,44,56,65,67,66,78,87,89,98,90,99]
data.sort()
n=len(data)

if n%2==0:
    median=data[n//2]
else:
    median=(data[n//2-1]+data[n//2])/2

print(median)
```

```python
###  Median if None Value   ###
data=[21,23,None,22,33,34,43,None,33,44,45,54,None,55,44,56,65,None,67,66,78,87,None,89,98,90,99]
clean=[x for x in data if x is not None]
clean.sort()
n=len(clean)

if n%2==0:
    median=clean[n//2]
else:
    median=(clean[n//2-1]+clean[n//2])/2

print(median)
```

```python
##  Median By Library  ##
import numpy as np
data=[21,23,None,22,33,34,43,None,33,44,45,54,None,55,44,56,65,None,67,66,78,87,None,89,98,90,99]
clean=[x for x in data if x is  not  None]
median=np.median(clean)
median
```

###   Std_Deviation  Methods  ###

```python
##  std_dev  simple  ##
data=[21,23,22,33,34,43,33,44,45,54,55,44,56,65,67,66,78,87,89,98,90,99]
mean=sum(data)/len(data)

var=sum((x-mean)**2 for x in data)/len(data)
std_dev=var**0.5
std_dev
```

```python
##   Std_dev  with None Values  ##
data=[21,23,None,22,33,34,43,None,33,44,45,54,None,55,44,56,65,None,67,66,78,87,None,89,98,90,99]
clean=[x for x in data if x is not None]
mean=sum(clean)/len(clean)

var=sum((x-mean)**2  for x in clean)/len(clean)
std_dev=var**0.5
std_dev
```

```python
##  Std_dev By Library  ##
import numpy as np
data=[21,23,22,33,34,43,33,44,45,54,55,44,56,65,67,66,78,87,89,98,90,99]
std_dev=np.std(data)
std_dev
```

###   Qrtile 

```python
##  In simple Formula ##
data=[21,23,22,33,34,43,33,44,45,54,55,44,56,65,67,66,78,87,89,98,90,99]
data.sort()
n=len(data)

if n%2==0:
    Q2=(data[n//2-1]+data[n//2])/2
else:
    Q2=data[n//2]


if n%2==0:
    lower_half=data[:n//2]
    upper_half=data[n//2:]
else:
    lower_half=data[:n//2]
    upper_half=data[n//2+1:]

n1=len(lower_half)

if n1%2==0:
    Q1=(lower_half[n1//2-1]+lower_half[n1//2])/2
else:
    Q1=lower_half[n1//2]

n2=len(upper_half)

if n2%2==0:
    Q3=(upper_half[n2//2-1]+upper_half[n2//2])/2
else:
    Q3=upper_half[n2//2]

IQR=Q3-Q1

print("Q1",Q1)
print("Q2",Q2)
print("Q3",Q3)
print("IQR",IQR)
```

```python
###  From Numpy  ##
import numpy as np
data=[21,23,22,33,34,43,33,44,45,54,55,44,56,65,67,66,78,87,89,98,90,99]
std_dev=np.std(data)
std_dev

```

```python
## From Statistics  Library ###
import statistics as sts
data=[21,23,22,33,34,43,33,44,45,54,55,44,56,65,67,66,78,87,89,98,90,99]
std_dev=sts.stdev(data)
std_dev
```
