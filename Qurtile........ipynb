###   Qrtile 

```python
##  In simple Formula ##
data=[21,23,22,33,34,43,33,44,45,54,55,44,56,65,67,66,78,87,89,98,90,99]
data.sort()
n=len(data)

if n%2==0:
    Q2=(data[n//2-1]+data[n//2])/2
else:
    Q2=data[n//2]


if n%2==0:
    lower_half=data[:n//2]
    upper_half=data[n//2:]
else:
    lower_half=data[:n//2]
    upper_half=data[n//2+1:]

n1=len(lower_half)

if n1%2==0:
    Q1=(lower_half[n1//2-1]+lower_half[n1//2])/2
else:
    Q1=lower_half[n1//2]

n2=len(upper_half)

if n2%2==0:
    Q3=(upper_half[n2//2-1]+upper_half[n2//2])/2
else:
    Q3=upper_half[n2//2]

IQR=Q3-Q1

print("Q1",Q1)
print("Q2",Q2)
print("Q3",Q3)
print("IQR",IQR)
```
